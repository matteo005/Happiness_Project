```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data/Clean Data/HappynessYearsCleanData.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df["Region"].value_counts()
```

```python
region_vs_gdp = df.groupby("Region")[[
    "2015GDP", "2016GDP", "2017GDP", "2018GDP", "2019GDP", "2020GDP", "2021GDP"
]].mean()

region_vs_gdp = region_vs_gdp.rename(columns={
    "2015GDP": "2015",
    "2016GDP": "2016", 
    "2017GDP": "2017", 
    "2018GDP": "2018", 
    "2019GDP": "2019", 
    "2020GDP": "2020",
    "2021GDP": "2021" 
})

region_vs_gdp
```

```python
region_vs_gdp.T.plot(rot=90, figsize=(10, 8))
plt.xlabel("Year")
plt.ylabel("GDP")
plt.title("Year vs GDP")
plt.legend(bbox_to_anchor=(1.4, 1), loc='upper right', ncol=1)
plt.show()
```

```python
region_vs_Happiness = df.groupby("Region")[["2015Happiness", "2016Happiness", "2017Happiness", "2018Happiness", "2019Happiness", "2020Happiness", "2021Happiness"]].mean()
region_vs_Happiness = region_vs_Happiness.rename(columns={
    "2015Happiness": "2015",
    "2016Happiness": "2016",
    "2017Happiness": "2017", 
    "2018Happiness": "2018", 
    "2019Happiness": "2019", 
    "2020Happiness": "2020",
    "2021Happiness": "2021" 
})

region_vs_Happiness
```

```python
region_vs_Happiness.T.plot(rot=90, figsize=(10, 8))
plt.xlabel("Year")
plt.ylabel("Happiness")
plt.title("Year vs Happiness")
plt.legend(bbox_to_anchor=(1.4, 1), loc='upper right', ncol=1)
plt.savefig("region_vs_happiness.png")
plt.show()
```
