```python
# pull in dependencies
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets

import seaborn as sns
```

```python
y_2015 = pd.read_csv('2015.csv')
y_2016 = pd.read_csv('2016.csv')
y_2017 = pd.read_csv('2017.csv')
y_2018 = pd.read_csv('2018.csv')
y_2019 = pd.read_csv('2019.csv')
y_2020 = pd.read_csv('2020.csv')
y_2021 = pd.read_csv('2021.csv')
y_2020
```

```python
#add year column with corresponding year of file
y_2015['Year']="2015"
y_2016['Year']="2016"
y_2017['Year']="2017"
y_2018['Year']="2018"
y_2019['Year']="2019"
y_2020['Year']="2020"
y_2021['Year']="2021"

#Cleanup column names to be uniform and the same across all DF before merging
y_2015 = y_2015.rename(columns={'Economy (GDP per Capita)':'GDP'})
y_2016 = y_2016.rename(columns={'Economy (GDP per Capita)':'GDP'})
y_2017 = y_2017.rename(columns={'Economy..GDP.per.Capita.':'GDP', 'Happiness.Score':'Happiness Score'})
y_2018 = y_2018.rename(columns={'Country or region':'Country','GDP per capita':'GDP', 'Score': 'Happiness Score'})
y_2019 = y_2019.rename(columns={'Country or region':'Country','GDP per capita': 'GDP',
                               'Score': 'Happiness Score'})
y_2020 = y_2020.rename(columns={'Country name':'Country', 'Explained by: Log GDP per capita': 'GDP',
                                'Generosity':'ng','Explained by: Generosity': 'Generosity','Ladder score': 'Happiness Score'})
y_2021 = y_2021.rename(columns={'Country name':'Country','Explained by: Log GDP per capita': 'GDP', 
                                'Generosity':'ng','Explained by: Generosity': 'Generosity','Ladder score': 'Happiness Score'})


df2015 = y_2015[['Country','Happiness Score','GDP','Freedom','Generosity','Year','Trust (Government Corruption)']]
df2016 = y_2016[['Country','Happiness Score','GDP','Freedom','Generosity','Year','Trust (Government Corruption)']]
df2017 = y_2017[['Country','Happiness Score','GDP','Freedom','Generosity','Year','Trust..Government.Corruption.']]
df2018 = y_2018[['Country','Happiness Score','GDP','Freedom to make life choices','Generosity','Year','Perceptions of corruption']]
df2019 = y_2019[['Country','Happiness Score','GDP','Freedom to make life choices','Generosity','Year','Perceptions of corruption']]
df2020 = y_2020[['Country','Happiness Score','GDP','Freedom to make life choices','Generosity','Year','Perceptions of corruption']]
df2021 = y_2021[['Country','Happiness Score','GDP','Freedom to make life choices','Generosity','Year','Perceptions of corruption']]
```
